<h2>Character History</h2>

As of the current version, Genshin Impact has over 85 characters! That's quite a lot in the collection. Below is the character count <br>
for both 4 Star Characters and 5 Star Characters

```python

rarities = ["4 Stars", "5 Stars"]

#Filter the table data based on 4 star and 5 star count

fourStars = genshin[genshin['Rarity'] == "4 Stars"]
fiveStars = genshin[genshin['Rarity'] == "5 Stars"]

f = plt.figure()
f.set_figwidth(9)
f.set_figheight(5)

#Find out how many of each rarity there are. HINT: Use len
plt.title("Number of Characters and their Rarities")
plt.bar(rarities,[len(fourStars),len(fiveStars)], color=('#9565DC','#F2B145'))

#Plot the data using a bar graph or a histogram

plt.xlabel("Rarities", fontsize = 13)
plt.ylabel("Character Count", fontsize = 13)
plt.show()
```

Throughout the game, there have been a total of 40 4 Stars and over 45 5 Stars. This goes to show that the game has been heavily leaning<br>
towards the creation of more 5 star units, which isn't a surprise. The more powerful units, the merrier!

```python
#Prepare the data filtering out the versions
versionNums = genshin['Version'].str[:1].astype(int)
majorVersions = ["1.X", "2.X", "3.X", "4.X"]
versionOne = genshin['Name'][versionNums == 1]
versionTwo = genshin['Name'][versionNums == 2]
versionThree = genshin['Name'][versionNums == 3]
versionFour = genshin['Name'][versionNums == 4]
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)

#Find out how many characters per each version there are. HINT: Use len
plt.title("Number of Characters per Version")
plt.bar(majorVersions, [len(versionOne),len(versionTwo),len(versionThree),len(versionFour)],color=('#85E6DA','#724DA6','#57C276','#3783DF'))

#Plot them out in the bar graph accordingly 
plt.xlabel("Major Versions")
plt.ylabel("Number of Characters")
plt.show()

```

Of course, Genshin Impact has gone through numerous version updates, and each of these major versions had <br>
a fair share of new characters to play around with. Throughout the game, the number of characters per<br>
major version are as follows:

<ul>
    <li>1.X: 34 Characters</li>
    <li>2.X: 17 Characters</li>
    <li>3.X: 16 Characters</li>
    <li>4.X: 18 Characters</li>
</ul>

It's pretty understandable for 1.X to have the highest character population since it's the initial <br>
release of the game. From there, they've been going consistent with the character additions, <br>
with 4.X having the most number of characters ever since the first major update.

Of course, each of these characters come from different regions, as the game heavily emphasizes on the massive<br>
regions and their respective environments. Just how many characters of a certain region are there?

```python
#List out the different regions
regions = ["Mondstadt", "Liyue", "Inazuma", "Sumeru", "Fontaine", "Snezhnaya"]

#Get the list of characters that come from a certain region
mondstadt = genshin[genshin['Region'] == "Mondstadt"]
liyue = genshin[genshin['Region'] == "Liyue"]
inazuma = genshin[genshin['Region'] == "Inazuma"]
sumeru = genshin[genshin['Region'] == "Sumeru"]
fontaine = genshin[genshin['Region'] == "Fontaine"]
snezhnaya = genshin[genshin['Region'] == "Snezhnaya"]

#Find out how many characters per each region there are. HINT: Use len
regionCount = [len(mondstadt),len(liyue),len(inazuma),len(sumeru),len(fontaine),len(snezhnaya)]

#Plot and visualize the data in a Pie Graph
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
plt.pie(regionCount, colors=('#85E6DA','#F5BC65','#724DA6','#57C276','#3783DF','#62CFE9'))
plt.title("Character Count per Region",fontsize=30)
plt.legend(regions)
plt.show()
```

```python
#Get the percentages of the character distributions by doing an operation on all the character counts
regionPercentage = np.round(( np.array(regionCount) / len(genshin) ) * 100,2)

#Print out the individual percentages as well as the character count
print("Percentages of characters per Region: \n\n========================================\n")
for x in range (len(regionPercentage)):
    print(regions[x] + ": " + str(regionPercentage[x]) + "%; with " + str(regionCount[x]) + " characters")
```

<h3>As we can see above, most characters hail from LIYUE, which makes sense given how Hoyoverse likes to pay homage to chinese culture.</h3><br>
The second largest population is from Mondstadt, which also makes sense, given how Hoyoverse has predominantly been focusing on Mondstadt and Liyue prior to the other regions<br><br>
The succeeding populations' sizes are 18.82% from Inazuma, 16.47% from Sumeru, 14.12% from Fontaine, and 2.35% from Snezhnaya.<br><br><br>
Surprisingly, Snezhnaya has more in-game playable characters than Natlan (which isn't even released yet), but that'll change once the next version comes out.


<h2>Characters of Specific Elements</h2>

The main focus of Genshin Impact's lore and gameplay is it's Elemental System, and how there are different characters wielding different elements,<br>
making for a diverse set of gameplay styles. As Genshin released characters overtime, characters of a specific element gradually began to increase.

```python
#List out the different elements
elements = ["Cryo", "Pyro", "Electro", "Hydro", "Anemo", "Geo", "Dendro"]

#Get the number of characters of a specific element in the table
elementCount = genshin["Vision"].value_counts()
print(elementCount)

#Initialize plotting out the bar graph
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)
plt.bar(elements,elementCount, color = ('#33E0FF','#FF7133','#7B45F9','#1480F3','#5FE9DD','#F3C127','#B1DE26'))
plt.title("Range of Characters per Element")
plt.xlabel("Element",fontsize = 15)
plt.ylabel("Character Count",fontsize = 15)
plt.show()
```

<h3>Most of the characters in the roster are either Pyro or Cryo. Eitherway, they have the same amount.<br><br></h3>

We then have Electro and Hydro characters being the next largest group. This makes sense considering<br>
the main Elemental Reaction cycle revoles around these 4. <br><br>

As for the other three, Anemo is the highest out of them seeing as how it plays a great role in either Crowd Control<br>
or enhancing certain Elemental Reactions via 'SWIRL' (which is Anemo + any of the main 4).<br><br>

Meanwhile Dendro and Geo have lower counts but have the same amount.<br>
Dendro is relatively new, but has been seen to have potential in the long run. <br>
Geo, while it's been there from the start, has issues in terms of its playability <br>

```python
#Initialize lists for 4 star characters and 5 star characters of a specific element respectively
fourStarElements, fiveStarElements = [],[]

#Append the counts, for characters of a specific element, to their respective lists
for x in range (len(elements)):
    fourStarElements.append(len((fourStars[fourStars["Vision"] == elements[x]]).value_counts()))
    fiveStarElements.append(len((fiveStars[fiveStars["Vision"] == elements[x]]).value_counts()))

#Initialize the plotting for the figure
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)
plt.plot(elements,fourStarElements, color=('#9565DC'))
plt.plot(elements,fiveStarElements, color=('#F2B145'))
plt.title("Range of 4 Stars and 5 Stars per Element")
plt.xlabel("Element",fontsize = 15)
plt.ylabel("Character Count",fontsize = 15)
plt.legend(['4 Star Characters','5 Star Characters'])
plt.text(0,4.5,"There are about \n" + str(sum(fourStarElements)) + " 4 Star Characters...",fontsize=8)
plt.text(0,3.7,"...and there are about \n" + str(sum(fiveStarElements)) + " 5 Star Characters",fontsize=8)
plt.text(0,3,"The other 2 are a special case",fontsize=8)
plt.show()
```

<h3>In terms of the 5 Stars, a lot of them belong to the Hydro element. Though Hydro was scarce back then, it's revitalizing itself by becoming meta.</h3><br>
However, that's not the same case with the 4 stars, as they're only 3 of them throughout the history of the game<br>
(Hoyoverse, please release more 4 Star Hydro Characters, ISTG :( :( :(     )<br><br>

As for 4 Stars, most of them belong to either Cryo or Pyro. This is all because of meta-wise reasons as well as the plot shown earlier.<br><br>

The lowest count for both rarities include Geo and Dendro. This ties back with the points elaborated earlier as to why their count is the lowest<br><br>

Overall, the distribution of characters from certain elements seem decent<br><br>

Although with the release of many regions, characters of a specific element have been prioritized for release.<br>
This poses the question of how many characters of a specific element hail from a certain region

```python
#Get the list of elements from the other part of the code
#Initialize 'regional elements as an empty multidimensional array
colorList = ['#85E6DA','#F5BC65','#724DA6','#57C276','#3783DF','#62CFE9']
regionElementList = []
#Use a for loop to add elements within a specific array, then add them into the main array
for x in range (len(regions)):
    regionAppend = genshin[genshin['Region'] == regions[x]]
    regionalElements = []
    for y in range(len(elements)):
        regionalElements.append(len(regionAppend[regionAppend["Vision"] == elements[y]]))
    regionElementList.append(regionalElements)

#Plot each of the arrays into their respective areas using a for loop
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(7)
for z in range (len(regions)):
    plt.plot(elements,regionElementList[z],color=(colorList[z]))


#Show the plot
plt.grid()
plt.title("Distribution of Character Elements across Regions",fontsize=25)
plt.xlabel("Elements", fontsize=15)
plt.ylabel("Character Count", fontsize=15)
plt.legend(regions)
plt.show()
```

<h3>Based on the graph above, here's how dominant an element is over a certain region:</h3>
<ul>
    <li>Mondstadt: Cryo</li>
    <li>Liyue: Pyro</li>
    <li>Inazuma: Electro</li>
    <li>Sumeru: Dendro</li>
    <li>Fontaine: Hydro and Cryo are tied</li>
    </li>Snezhnaya: Not enough to say</li>
</ul>

<h4>Mondstadt had predominantly Cryo characters, while Liyue had predominantly Pyro characters. </h4><br>
There wasn't any particular reason as to why, but they chose to emphasize it upon release<br><br>

Inazuma had predominantly Electro characters primarily because it was fitting to Inazuma's <br>
element, and that Electro characters were a bit scarce back then. <br><br>

Sumeru had predominantly Dendro characters primarily because it fit Sumeru's element, which<br>
made it the perfect moment to finally release Dendro: Genshin's missing element.<br><br>

Fontaine's majority were tied between Hydro and Cryo. Hydro made sense, but Cryo was<br>
more of something that just happened to be there. It was welcome nonetheless<br><br>

There's not enough data to say something abut Snezhnaya, so we'll leave it like that



<h2>Characters Wielding Different Weapons</h2>

Of course, each of the characters wield their own weapons. Sometimes, these weapon choices offer<br>
a chance to exploire different combat styles and strategies



```python
#Initialize the list of weapons
weapons = ["Sword","Bow","Catalyst","Polearm","Claymore"]

#Count how many characters, wielding a specific weapon, therem are
weaponCount = genshin["Weapon"].value_counts()

#Initialize the bar plot
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)
plt.bar(weapons,weaponCount,color=('#1D90CE','#D52C2C','#9A68F1','#F87B3C','#F1B168'))
plt.title("Range of Characters per Weapon")
plt.xlabel("Weapon",fontsize = 15)
plt.ylabel("Character Count",fontsize = 15)
plt.show()
```

<h3>Most of the characters are sword users, while the least amount of characters are claymore users</h3><br>

The reast of the weapon types (like bows, catalysts, and polearms) have a more or less equal amount.<br><br>

Though, it should be noted that weapon types don't have much of an influence on the meta, so the number<br>
of characters wielding a specific weapon don't hold much weight here.

However, just like before, 5 stars and 4 stars wield different types.

```python
#Initalize the list of 4 stars and 5 stars wielding a specific weapon
fourStarWeapons, fiveStarWeapons = [],[]

#Append each of them to their respecitve lists
for x in range (len(weapons)):
    fourStarWeapons.append(len((fourStars[fourStars["Weapon"] == weapons[x]]).value_counts()))
    fiveStarWeapons.append(len((fiveStars[fiveStars["Weapon"] == weapons[x]]).value_counts()))

#Initially plot out the figure
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)

plt.plot(weapons,fourStarWeapons, color=('#9565DC'))
plt.plot(weapons,fiveStarWeapons, color=('#F2B145'))
plt.title("Range of 4 Stars and 5 Stars per Weapon")
plt.xlabel("Weapon",fontsize = 15)
plt.ylabel("Character Count",fontsize = 15)
plt.legend(['4 Star Characters','5 Star Characters'])
plt.text(0,6.3,"There are about \n" + str(sum(fourStarWeapons)) + " 4 Star Characters...",fontsize=8)
plt.text(0,5.6,"...and there are about \n" + str(sum(fiveStarWeapons)) + " 5 Star Characters",fontsize=8)
plt.text(0,5,"The other 2 are a special case",fontsize=8)
plt.show()
```

The relationship between weapon types of a certain rarity is rather interesting here:<br><br>

Most 5 Star Characters are Sword users, while the least amount of them are claymore users.<br>
Subsequently, most 4 Star Characters are Claymore users, while the least amount of them are sword users.<br>
The amount of bows and polearms are equal, while the amount of catalyst users between the two are spread out.<br><br>

In order words, the number of characters, wielding a specific weapon, for both 4 stars and 5 stars...<br>
<h3>...is INVERSELY PROPORTIONAL</h3><br>

Aside from that, different characters of different elements wield diffferent weapons (boy, that's a lot of differences).<br>
Though just how spread out are the two from each other? How many characters of Element X wield Weapon Y?

```python
#Initialize the list containing the number of characters of a specificv element...
#...wielding a certain weapon

weaponElementList = []
colorListTwo = ['#1D90CE','#D52C2C','#9A68F1','#F87B3C','#F1B168']
#Use a for loop to add elements within a specific array, then add them into the main array
for x in range (len(weapons)):
    weaponAppend = genshin[genshin['Weapon'] == weapons[x]]
    weaponElements = []
    for y in range(len(elements)):
        weaponElements.append(len(weaponAppend[weaponAppend["Vision"] == elements[y]]))
    weaponElementList.append(weaponElements)

#Plot each of the arrays into their respective areas using a for loop
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(7)
for z in range (len(weapons)):
    plt.plot(elements,weaponElementList[z],color=colorListTwo[z])


#Show the plot
plt.grid()
plt.title("Distribution of Character Elements with Different Weapons",fontsize=25)
plt.xlabel("Elements", fontsize=15)
plt.ylabel("Character Count", fontsize=15)
plt.legend(weapons)
plt.show()
```

<h3>Based on the graph above, the most amount of weapons per element are as follows: </h3>
<ul>
    <li>Cryo: Sword</li>
    <li>Pyro: Polearm</li>
    <li>Hydro: Sword, Catalyst</li>
    <li>Electro: Claymore, Sword, and Bow</li>
    <li>Anemo: Sword</li>
    <li>Geo: Claymore</li>
    <li>Dendro: Everything except Claymore</li>
</ul>

Elements like Cryo, Pyro, Geo, and Anemo have a clearly predominant weapon amongst their characters<br>
while Hydro, Dendro, and Electro are well distributed (at least with some weapons)<br><br>

As stated before, the number of characters wielding a certain weapon have nothing to do with the impact<br>

<h2>Gender Range of Characters</h2>

Of course, the game's entire roster of characters consists of a wide arrange of character models and body types. However, considering the demographic<br>
Hoyoverse is aiming for, it's possible for one gender to have more characters than the other. Below is the number of male and female characters<br>
that are currently present in the game. 

```python
#Filter out the dataset for characters that are male and female
male = genshin[genshin['Height'].str[-4:] == "Male"]
female = genshin[genshin['Height'].str[-6:] == "Female"]

print(len(male))
print(len(female))
#Prepare the plot
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(5)
plt.bar(["Male","Female"],[len(male),len(female)], color=('#4599F2','#DF6DE8'))

plt.title("Distribution of Characters per Gender")
plt.xlabel("Gender")
plt.ylabel("Character Count")
plt.show()
```

<h3>As you can see, there is a big gap between the amount of male and female characters.</h3>
There are over 30 Male characters while there are over 55 Female Characters, which as you can see, is a huge gap.<br><br>
Many people have expressed their thoughts on how they think the game could use more Male characters, but<br>
it seems like the game is catering towards more female characters in order to rope in more of a specific audience.<br>
EIther that or they just really have a preference when it comes to designing female characters.<br><br>
However, just like the other cases, these characters are divided by rarity.

```python
#Get the datasets of 4 Star Characters and 5 Star Characters belonging to a specific gender
fourStarMale = male[male["Rarity"] == "4 Stars"]
fiveStarMale = male[male["Rarity"] == "5 Stars"]
fourStarFemale = female[female["Rarity"] == "4 Stars"]
fiveStarFemale = female[female["Rarity"] == "5 Stars"]

print(len(fourStarMale))
print(len(fourStarFemale))
print(len(fiveStarMale))
print(len(fiveStarFemale))

#Prepare the plot
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(5)

plt.bar(["Male","Female"],[len(fiveStarMale),len(fiveStarFemale)], color=('#F2B145'))
plt.bar(["Male","Female"],[len(fourStarMale),len(fourStarFemale)], color=('#9565DC'))
plt.title("Distribution of 4 Stars and 5 Stars per Gender")
plt.xlabel("Gender")
plt.ylabel("Character Count")
plt.show()
```

<h3>Even when delving into character rarity, female characters at a higher amount than male characters</h3>
For the 4 stars, about 13 are male and 27 are female...<br>
..while for the 5 stars, 17 are male and 28 are female. <br>
Again, it indicates to us that female characters are highly focused on in this game.<br><br>
